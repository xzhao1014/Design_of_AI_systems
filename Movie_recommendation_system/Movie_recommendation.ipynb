```python
import pandas as pd
import numpy as np
from pandas import DataFrame
ratings = pd.read_csv(r'user_reviews.csv')
```

```python
ratings = ratings.drop(["Unnamed: 0"], axis = 1).set_index(["User"]).astype(int)
#ratings = ratings.drop(ratings.index[200:500],axis=0)
```

```python
#The following commands are to shrink the original data set
#delete the users with least ratings on the movies
#duser = ratings.sum(axis=1).sort_values(ascending=True)[:550].index.values.tolist()

#delete the users from the bottom of the list
ratings = ratings.drop(ratings.index[300:],axis=0)

#delete movies with least ratings from the users
ditem = ratings.sum(axis=0).sort_values(ascending=True)[:1000].index.values.tolist()
ratings = ratings.drop(ditem, axis = 1)#.drop(duser, axis = 0)
ratings
```

```python
mcors = np.corrcoef(ratings, rowvar=1)
mcors = mcors*0.5+0.5
mcors = pd.DataFrame(mcors, columns=ratings.index, index=ratings.index)
mcors
```

```python
def cal_score(matrix,mcors,item,user):
    totscore = 0
    totsims = 0
    score = 0
    if matrix[item][user]==0:
        for muser in ratings.index:
            if matrix[item][muser]==0:
                continue
            if pd.isnull(mcors[user][muser]):
                continue
            else:
                totscore += matrix[item][muser]*mcors[user][muser]
                #totsims += mcors[user][muser]
                totsims = totsims + 1
        if totsims == 0:
            score = matrix[item][user]
        else:
            score = totscore/totsims
    else:
        score = matrix[item][user]
    return score
```

```python
def cal_matscore(matrix,mcors):
    score_matrix = np.zeros(matrix.shape)
    score_matrix = pd.DataFrame(score_matrix, columns=matrix.columns, index=matrix.index)
    for mitem in matrix.columns:
        for muser in matrix.index:
            score_matrix[mitem][muser] = cal_score(matrix,mcors,mitem,muser)
    return score_matrix
```

```python
def recommend(matrix,score_matrix,user,n):
    user_ratings= matrix.loc[user]
    not_rated_item = user_ratings[user_ratings==0]
    recom_items = {}
    for item in not_rated_item.index:
        recom_items[item]=score_matrix[item][user]
    recom_items = pd.Series(recom_items)
    recom_items = recom_items.sort_values(ascending=False)
    return recom_items[:n]
```

```python
score_matrix = cal_matscore(ratings,mcors)
```

```python
for i in range(5):
    user = input(str(i)+" please input the user name:")
    print(recommend(ratings,score_matrix,user,5))
```
